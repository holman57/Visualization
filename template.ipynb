```python
import pandas as pd
import matplotlib.pyplot as plt

fig, ax = plt.subplots(dpi=800)
df = pd.DataFrame({
    'Name': ['Abby', 'Brian', 'Curtis', 'Dev', 'Emily'],
    'Years working in AI': [5, 5, 10, 15, 30],
})
ax.set_xlabel("Names")
ax.set_ylabel("Years working in AI")
ax.set_title("Engineers' AI Experience")
ax.bar(df['Name'], df['Years working in AI'])
fig.savefig("bar.png")
plt.show()
plt.close(fig)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(dpi=800)
years = np.array(['2001', '2002', '2003', '2004'])
sample = np.array([3, 8, 1, 10])
ax.set_xlabel("Year")
ax.set_ylabel("Population")
ax.set_title("Population Trends Across Time")
ax.plot(years, sample)
fig.savefig("line.png")
plt.show()
plt.close(fig)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(dpi=800)
x = np.array([5, 10, 16, 15, 22, 28, 64, 23])
y = np.array([11, 18, 2, 3, 7, 8, 45, 32])
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Relationship between x and y")
ax.scatter(x, y)
fig.savefig("scatter.png")
plt.show()
plt.close(fig)
```
